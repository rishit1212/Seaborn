# Relational/Statistical Plots

Relational Plots are used to perform multivariate analysis

## Scatter Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

## Plot a Scatter Plot b/w total_bill and tip

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
sns.relplot(x='total_bill',hue='smoker',y='tip',data=data) #categorical data for hue parameter
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data) #categorical data for kind parameter
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',data=data) #continous data for sizr parameter
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),data=data) #specifies area
```

### Using scatterplot function

```python
sns.scatterplot(x='total_bill',hue='smoker',y='tip',data=data)
```

## Plotting with Categorical Data

### catplot[Figure Level Function]

## 1.Categorical Scatterplots

```python
data.head()
```

```python
sns.catplot(x='day', y='tip', kind='strip',data=data) #default kind = stri0
```

```python
sns.swarmplot(x='day',y='tip',data=data)
```

## 2.Categorical Distribution plots

### Boxplots

```python
#Univariate
sns.boxplot(data['total_bill'])
```

```python
#Bivariate
sns.catplot(x='day',y='total_bill',kind='box',data=data)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=data)
```

### Violinplots

##### The advantage of violinplot over boxplot is that aside from above mentioned statistics it also shows the entire distibution of the data.

##### Violin plots are often used to compare the distribution of a given variable across some categories.

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=data)
```

## 3.Categorical estimate plots

### Barplot

```python
#These graphs also gives info about Cantral Tendancies
#Default behaviour of bar plot is to calculate mean of the numerical value
```

```python
#Que Find out mean value of tip for smokers and non-smokers
```

```python
# We want info of numerical data with respect to categorical data
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=data)
```

```python
#Error Bars showing confidence interval for calculation of mean
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',estimator=np.var,hue='sex',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',estimator=np.std,hue='sex',data=data)
```

## 4.Categorical Categorical plots

### Heatmaps

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='summer')
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='winter')
```

### Clustermap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,row_cluster=False,metric='correlation')
```

## Pairplots

Pairplots are used to plot multiple bivariate distribution in a dataset.This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame.By default,it also draws the univariate distribution of each variable on the diagonal Axes.

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

## Distplots

Distplots are used to do univariate analysis on a continous/numerical data.

It combines 3 things in one graph:

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],bins=10,kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

##### How to use Distplot  

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
